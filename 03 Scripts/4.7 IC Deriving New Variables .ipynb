# Using If Statement

```python
#Define price label function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply price label on dataframe
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# check the max price for products
df['prices'].max()
```

Question 4: Identify the busiest hours of the day. Label as  “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# Check frequency of values in the "order_hour_of_day" column 
df_ords_prods_merged['order_hour_of_day'].value_counts()
```

We can group the above order_hour_of_day into the following categories (as per the clients instructions) = "Most orders,” >= 2000000 “Average orders 2000000 <= 500000,” and “Fewest orders > 500000.

```python
# Create new if-statement for-loop for new requirements
hour_of_day = []

for value in df_ords_prods_merged['order_hour_of_day']:
  if value in [10, 11, 14, 15, 13,12, 16, 9, 17]:
    hour_of_day.append('Most orders')
  elif value in [8, 18, 19, 20, 7, 21, 22]:
    hour_of_day.append('Average orders')
  else:
    hour_of_day.append('Fewest orders')
```

```python
# Add result to new column in dataframe
df_ords_prods_merged['busiest_period_of_day'] = hour_of_day
```
